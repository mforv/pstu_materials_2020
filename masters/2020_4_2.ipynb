```python
#  
class MyFunctor:
    def __init__(self, k):
        self.k = k
        
    def __call__(self, n):
        return self.k * n
```

```python
functor = MyFunctor(9)
```

```python
functor(5)
```

```python
# 
def currying_mult(x):
    def inner_mult(y):
        return simple_mult(x, y)
    return inner_mult
```

```python
currying_mult(9)(5)
```

```python
cr = currying_mult(9)
```

```python
cr(6)
```

```python
cr(5)
```
